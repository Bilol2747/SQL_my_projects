##Yuklab olingan omborga (chinook.db) ga ulanamiz

```python
db=sql.connect("chinook.db")
cursor=db.cursor()
```

### DB tarkibidagi jadvallarni ko'ramiz

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables=cursor.fetchall()
for table in tables:
  print(table[0])
```
